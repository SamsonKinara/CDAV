```python
def greeter_3(lang):
  greetings={"fr":"bonjour",
             "sw":"we mzee",
             "en-br":"hi",
             "jp":"konichiwa",
             "en-us":"howdy"}
  return greetings[lang]
```

```python
print(f"{greeter_3('en-us')} john")
```

```python
def authenticator(password):
  defined_pass="secret"
  if password == defined_pass:
    return True
  else:
    return False
```

```python
passw = input("Enter a password ")
if authenticator(passw):
  print("you are logged in")
else:
  print("Wrong password")
```

```python
def sum_diff(x,y):
  s=x+y
  d=x-y
  return(s,d)
```

```python
sum_of_both, diff_of_both = sum_diff(100,50)
```

```python
print(sum_of_both)
```
