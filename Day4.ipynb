In this exercise you will create a program that reads a letter of the alphabet from the user. If the user enters a, e, i, o or u then your program should display a message indicating that the entered letter is a vowel. If the user enters y then your program should display a message indicating that sometimes y is a vowel, and sometimes y is a consonant. Otherwise your program should display a message indicating that the letter
is a consonant.


```python
letter=input("Enter letter: ")
if letter=="y":
  print(" sometimes y is a vowel, and sometimes y is a consonant")
elif letter==("a")or letter==("e")or letter==("i")or letter=="u":
  print(f"{letter} is a vowel")
else:
  print(f"{letter} is a consonant")
```
